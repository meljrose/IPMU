# Introduction

This project deals with black holes, AGN, and dark matter halos in/around galaxy clusters. 

Specifically, we investigate two questions: 
1. Does the baryonic accretion rate of black holes in cD galaxies relate to the dark matter halo accretion rate for galaxy clusters? 
2. Does the occurence of AGN relate to the accretion rate of black holes or dark matter halos? 

To answer the first question, we need to compare the accretion rate of black holes and the accretion rate of dark matter halos in many galaxy clusters. To answer the second question, we need to additionally detect if AGN are present in the galaxy clusters. 

First, we measure the dark matter halo accretion rates in [Section 2](#Dark-Matter-Halo-Accretion-Rate) by determining the splashback radii of our galaxy cluster sample.

# Dark Matter Halo Accretion Rate

For dark matter halos, the traditional halo edge is taken to be at the distance at which the halo encloses 200 times the mean matter density of the Universe ($R_{200m}$). For this project, we instead use the splashback radius ($R_{sp}$). 

The splashback radius (also known as the last density caustic or turnaround radius) is a sharp dark matter halo edge that corresponds to the location of the first orbital apocenter of satellite galaxies after their infall <a name="ref-2"/>[(More et al. 2016)](#cite-more:2016). We check whether observations of splashback radii can act as a proxy for mass assembly of galaxy clusters by comparing two galaxy cluster samples with differing assembly histories. For dark matter halos of the same mass, a large mass accretion rate should result in a smaller splashback radius, since its deeper halo potential well has a smaller value for its apocenter. 

We hypothesize that there is such a connection between the member galaxy distributions and mass assembly of these clusters.

## Method

We begin by matching our two high-c and low-c photometric catalogs with a few other catalogs: 
* <a href='http://wwwmpa.mpa-garching.mpg.de/SDSS/DR7/'>MPA-JHU emission line analysis stellar masses and SFR </a>
* <a href="http://www.sdss.org/dr12/spectro/galaxy_portsmouth/">Portsmouth Emission-Line Kinematics datamodel</a>
* <a href="http://heasarc.gsfc.nasa.gov/W3Browse/radio-catalog/nvss.html">NVSS - NRAO VLA Sky Survey Catalog</a>
* <a href="http://sundog.stsci.edu/first/catalogs.html">FIRST Survey Catalogs</a>

This was done using the best match Sky algorithm in <a href="http://www.star.bris.ac.uk/~mbt/topcat/">TOPCAT</a> for the RA and DEC entries, with a tolerance of 1 arcsecond. These catalogs additionally give us stellar masses, SFR, spectral line information, BPT classification, and radio emission data.

We then search for nearby galaxies with spectroscopic data within 10 Mpc of each galaxy in the catalogs. We add the number of satellite galaxies found into bins as a function of radius from the central galaxy. Each spectra is tagged as a Seyfert, Liner, Composite, etc., for each galaxy. We compare the type of galaxy surrounding each central galaxy with the galaxy cluster's dark matter halo accretion rate. 

We convert from RA/DEC in degrees to cartesian coordinates in radians on a unit sphere and then apply a redshift cut based on radial velocity.  We use a KDTree for nearest neighbor matches. We use each photometric galaxy as a central, "seed," galaxy to search in a spherical radius around. We use the default cosmological parameters found in AUM <sup><a href="#fn3" id="ref3">3</a></sup>, a flat ΛCDM cosmological model with matter density parameter Ωm = 0.27 and the Hubble parameter h = 0.7, for calculating radii at various redshifts.

```python
# %%cython
# Searching Functions


import numpy as np
import pandas as pd
from astropy.table import Table
import pyfits
from scipy import spatial
import cosmology as cc
a = cc.cosmology()

def fits_to_df(fits_file, colnames):
    
    df = pd.DataFrame(columns=colnames, dtype=object)
    hdulist = fits.open(fits_file, memmap=True)
    tbdata = hdulist[1].data
    
    for col in colnames:
        try:
            df[col] = tbdata[col]
        except:
            print("column '{0}' not in file {1} \n".format(col, fits_file))
        
    hdulist.close()
    return(df)

def get_colnames(fits_file):
    hdulist = fits.open(fits_file, memmap=True)
    tbdata = hdulist[1].data
    colnames = tbdata.names
    hdulist.close()
    return(colnames)
        
def load_cols(fits_file, df, colnames, indices):
    '''    indices = [1,2,3]
    colnames = ['NTMASS','SM_P84']
    '''
    hdulist = fits.open(fits_file, memmap=True)
    tbdata = hdulist[1].data

    for col in colnames:
        if col not in df.columns:
            df[col] = np.nan

        for index in indices:
            tempBool = np.isnan(df.loc[[index], col].values[0])
            if tempBool:
                df.loc[[index], col] = pd.Series([tbdata[index][col]], index = [index])
    hdulist.close()
    return(df)

def tocartesian(ra_in_deg, dec_in_deg):
    """(x,y,z) = tocartesian(ra_in_deg, dec_in_deg)
    
    Convert RA and DEC coordinates given 
    in degrees into Cartesian points (x,y,z)
    on unit circle
    """
    
    ra = np.radians(ra_in_deg)
    dec = np.radians(dec_in_deg)
    x = np.cos(ra) * np.cos(dec)
    y = np.sin(ra) * np.cos(dec)
    z = np.sin(dec)
    
    return((x,y,z))

def df_to_tree(df):
    # spatial.cKDTree is the slow part
    coordinates = list(map(tocartesian,df['RA'],df['DEC']))
    tree = spatial.cKDTree(coordinates)
    return(tree, coordinates)


def angular_search_radius(radius_in_mpc,z):
    """
    angular_radius = angular_search_radius(radius_in_mpc,z)
    
    Return the comoving distance as a function of redshift, using default cosmological parameters
    in AUM (https://github.com/surhudm/aum/blob/master/README.md)
    """
    return(radius_in_mpc/a.Dcofz(z))


def redshift_cut(df,angular_matched_indices, z, radial_velocity):
    """
    cut_indices = redshift_cut(df,angular_matched_indices, z, radial_velocity)
    
    Return indices that are constrained by np.abs((C*delta_z)/(1+z)) < radial_velocity
    """
    
    # if there were no angular matches, then there is nothing to cut out
    if len(angular_matched_indices) == 0:
            return([])
        
    # retrieve redshifts for the matched galaxies    
    z_arr = [df.iloc[k]['REDSHIFT'] for k in angular_matched_indices]
    C = 3*10**5 # km/s
    
    # if the cut quantity is less than the radial velocity, keep the galaxy
    cut_mask = []
    for z_i in z_arr:
        delta_z = z - z_i
        cut = np.abs((C*delta_z)/(1+z))
        cut_mask.append(cut < radial_velocity)
    cut_indices = [index for (index, mask) in zip(angular_matched_indices, cut_mask) if mask]
    
    
    # for every few iterations, print out the variables so I can debug
    rand = np.random.binomial(1, 0.001)
    if rand > 0:
        print("for rv {0} and z {4:.2f}, {1}/{2} galaxies are kept; " 
              "last cut value is {3:.2f}, last delz value is {5:.2f}".format(
               radial_velocity, len(cut_indices),len(angular_matched_indices), cut,z, delta_z))
    
    return(cut_indices)

def getneighbors(seed_df, tree_df,
                 tree, mpc_search_radius, radial_velocity_limit):
    """
    indices, distances, types = getneighbors(seed_df, tree_df, tree_coordinate_list, 
                                             tree, mpc_search_radius, radial_velocity_limit)
                 
    Nearest neighbor query for a cKDTree, using a spherical search radius that depends on redshift
    and cosomological parameters, returning index and distance. Also retrieves the surrounding galaxy 
    types listed in the input table. Returns an array of arrays so if you were interested in one central
    galaxy in particular, you could find the surrounding galaxies for that single central galaxy index.
    
    Includes nearest neighbors whose distance is ~0.0
    
    Input
    ------
    seed_df                the pandas dataframe of galaxies you want to search a spherical area around
    tree_df                the pandas dataframe of galaxies populated in your search tree
    tree_coordinate_list   a list of cartesian coordinates for the galaxies populated in your search tree
    tree                   the cKDETree you want to search
    mpc_search_radius      the radius in Mpc for your search sphere around your seed galaxy
    radial_velocity_limit  the radial velocity variance cut (see help(redshift_cut))
    
    Output
    ------
    indices                array of arrays of indicies of the matched galaxies in the tree_df per seed galaxy
    distances              array of arrays of distances of the matched galaxies in the tree_df per seed galaxy
    types                  array of arrays of galaxy types (Seyfert, LINER, etc.) of the matched galaxes
                                                            in the tree_df per seed galaxy
    
    """
    
    col = 'tree_indices'
 
    for index,row in seed_df.iterrows():
        # select seed/central galaxy to search spherically around
        seed = row
        seed_z = float(seed['REDSHIFT'])
        seed_coordinates = tocartesian(seed['RA'],seed['DEC'])

        # retrieve angularly matched coordinates in the tree
        # also apply a redshift cut for variance above some radial velocity
        angular_matched_indices = tree.query_ball_point(seed_coordinates,
                                                        angular_search_radius(mpc_search_radius,seed_z))
        redshift_and_angular_matched_indices = redshift_cut(tree_df, 
                                                            angular_matched_indices, 
                                                            seed_z, radial_velocity_limit)

        seed_df.loc[[index], col] = pd.Series([redshift_and_angular_matched_indices], index = [index])
    
    return(seed_df)


def get_distances(seed_df, tree_df, stripnearest=False):
    distances = []
    for index,row in seed_df.iterrows():
        tree_indices = row['tree_indices']
        if len(tree_indices) == 0:
            distances.append([])
            continue

        z = float(row['REDSHIFT'])
        seed_point = tocartesian(row['RA'],row['DEC'])

        point_arr = [tocartesian(tree_df.loc[index]['RA'],tree_df.loc[index]['DEC']) for index in tree_indices]

        points = np.tile(np.asarray(seed_point), (len(point_arr),1))
        dists = list(map(np.linalg.norm, points - point_arr))

        # multiply distance in radians by comoving distance to get it in Mpc
        d_comoving = a.Dcofz(z)
        dists = [d_comoving*d for d in dists]

        if stripnearest:
            min_distance = 0.003 # in Mpc, this is about the 1/10th radius of the MW
            zeros = [i for i,x in enumerate(dists) if dists[i] > min_distance]
            dists = [dists[i] for i in zeros]
            tree_indices = [tree_indices[i] for i in zeros]
            seed_df.loc[[index], 'tree_indices'] = pd.Series([tree_indices], index = [index])
        distances.append(dists)
        
    return(distances)




def get_dists_of_type(string, type_list, distances_list):
    """
    dists = get_dists_of_type(string, type_list, distances_list)
    
    Retrieve the distance for a certain galaxy type (Seyfert, LINER, etc.)
    """
    indices = [i for i, x in enumerate(type_list) if x == string]
    dists = [distances_list[i] for i in indices]
    
    return(dists)


def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return(idx) #array[idx]

def get_HaHb_fluxes_for_plotting(df): 
    lines = [5007,4861,6583,6562]
    
    # retrieve the line values from the spectral galaxies dataframe
    fluxes = [retrieve_line_fluxes(lines, row, index) for index, row in df.iterrows()]
    O3,HBeta, N2, HAlpha = [np.asarray(flux) for flux in zip(*fluxes)]
    print("all done")

    # organize plotting variables
    type_list = df['BPT'].values

    mass_list = df['SM_AVG'].values
    x = np.log10(N2/HAlpha)
    y = np.log10(O3/HBeta)
    
    return(x,y,type_list, mass_list)

def retrieve_line_fluxes(which_lines, df_row, j):
    fluxes = []
    for i, line in enumerate(which_lines):
        index = find_nearest(df_row['LAMBDA'],line)
        flux = df_row['FLUX'][index]
        fluxes.append(flux)
        if j < 1:
            # gross string manipulation to double check line indexing
            split = df_row['NAME'].split()
            names = split[:5]
            names.append(split[5][:7])
            names.append(split[5][7:14])
            names.append(split[5][14:])
            names = np.concatenate([names,split[6:]])
            print("Searched {0}; found {1} at {2}".format(
                    line, names[index], df_row['LAMBDA'][index]))
    return(fluxes)

def get_string_set(typelist, full=True):
    string_set = list(set(typelist))
        # fix typos in the table
    for i in np.arange(len(typelist)):
        if typelist[i] == 'Star Fo':
            typelist[i] = 'Star Forming'
            
    for i in np.arange(len(typelist)):
        if typelist[i] == 'Composi':
            typelist[i] = 'Composite'
            
            
    if full:        
        bad_strings=['BLANK        ','             ','Star Fo', 'BLANK','Composi', ' ','']
    else:
        bad_strings=['BLANK        ','             ','Star Fo', 'BLANK','Composi', ' ','','Seyfert/LINER']
    for bad_string in bad_strings:
        if bad_string in string_set: 
            string_set.remove(bad_string)
    return(string_set)

def get_ndensity(data, bins):
    which_bins = np.digitize(data, bins)
    total_count = np.bincount(which_bins)
    bbins = np.concatenate([[0], bins])
    number_density = []
    for i in np.arange(len(bbins)-1):
        j = i+1
        number = total_count[i]
        area = np.pi*(bbins[j]**2-bbins[i]**2)
        number_density.append(number/area)
    return(number_density)

def get_fraction(data_arr, bins):
        which_bins = np.digitize(np.concatenate(data_arr), bins)
        total_count = np.bincount(which_bins)
        #print('Binning Info:  {0}'.format(scipy.stats.describe(total_count)))
        bar_arr = []
        for i in np.arange(len(data_arr)):
            digitized = np.bincount(np.digitize(data_arr[i], bins))
            bar_data = [digitized[i]/total_count[i] for i in np.arange(len(total_count))]
            bar_arr.append(bar_data)
        return(bar_arr)
    
def get_colnames(fits_file):
    hdulist = fits.open(fits_file, memmap=True)
    tbdata = hdulist[1].data
    colnames = tbdata.names
    hdulist.close()
    return(colnames)

def get_tree_column(seed_df, tree_df, column, load_new=False, fits_file='../data/merged1.fits'):
    
    if load_new: 
        colnames = get_colnames(fits_file)
        if column in colnames:
            indices = list(set(np.concatenate(seed_df['tree_indices'].values)))
            tree_df = load_cols(fits_file, tree_df, [column], indices)
        else:
            print("'{0}' does not exist in {1}".format(column, fits_file))
            return(None)
            
        
    if column not in tree_df.columns:
        print("'{0}' not loaded. Checking if it exists in {1}".format(column, fits_file))
        colnames = get_colnames(fits_file)
        if column in colnames:
            print("'{0}' exists in {1}. use load_new=True".format(column, fits_file))
        else:
            print("'{0}' does not exist in {1}".format(column, fits_file))
        return(None)
        

    column_arr = []
    for index,row in seed_df.iterrows():
        tree_indices = row['tree_indices']
        if len(tree_indices) == 0:
            column_arr.append([])
            continue
        columns = [tree_df.loc[i][column] for i in tree_indices]
        column_arr.append(columns)
    return(column_arr)
```

```python
def load_cols(fits_file, df, colnames, indices):
    '''    indices = [1,2,3]
    colnames = ['NTMASS','SM_P84']
    '''
    hdulist = fits.open(fits_file, memmap=True)
    tbdata = hdulist[1].data

    for col in colnames:
        if col not in df.columns:
            df[col] = np.nan
        for index in indices:
            
            # check if index already exists
            
            val = df.loc[[index], col].values[0]

            
            try: 
                tempBool = pd.isnull(val).all()
            except:
                tempBool = pd.isnull(val)
            

            if tempBool:

                df.loc[[index], col] = pd.Series([tbdata[index][col]], index = [index])
            
                
    hdulist.close()
    return(df)
```

We check the SFR using these formulae:

$$
E_{neb} (B-V) = \frac{2.5}{k_{H_\beta} - k_{H_\alpha}} \log_{10}\left(\frac{H_\alpha / H_\beta}{2.86}\right)
$$

$$
f_{int} = f_{obs} \cdot 10^{k_{H_\alpha} \cdot E_{neb}(B-V)}
$$

$$
L_{H_\alpha} = 4 \pi D^2 f_{int}
$$

```python
def e_neb(halpha, hbeta, k_halpha, k_hbeta):
    return(2.5/(k_hbeta - k_halpha)*np.log10((halpha-hbeta)/2.86))

def f_int(f_obs, halpha, hbeta):
    k_halpha = 3.325 
    k_hbeta = 4.598
    a_halpha = k_halpha*e_neb(halpha, hbeta, k_halpha, k_hbeta)
    return(f_obs * 10**(a_halpha))

def lum_halpha(f_obs, halpha, hbeta, z):
    flux = f_int(f_obs, halpha, hbeta)
    d_comoving = a.Dcofz(z)
    return(4*np.pi*d_comoving*flux)

print(lum_halpha(1000.0, 11.0, 10.0, 10.0))
```
